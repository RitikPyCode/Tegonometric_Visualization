# Import Important Libraries 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.sin(30)
```

```python
np.cos(45)
```

```python
np.tan(60)
```

```python
x_sin=np.arange(0, 3*np.pi, 0.1)
```

```python
x_sin
```

```python
y_sin=np.cos(x_sin)
```

```python
y_sin
```

```python
plt.plot(x_sin, y_sin)
plt.show()
```

```python
plt.plot(x_sin, y_sin, color="g")
plt.title("Tegnomatric_Calcutions ")
plt.xlabel("x-Axies")
plt.ylabel("y-Axies")
plt.grid(True)
plt.show()
```

```python
y_tan=np.tan(x_sin)
plt.plot(x_sin, y_tan)
plt.show()
```

```python
y_tan=np.tan(x_sin)
plt.plot(x_sin, y_tan, color="r")
plt.title("Modern Graph")
plt.xlabel("X-Axies")
plt.ylabel("y-Axies")
plt.grid(True)
plt.show()
```

```python
y_tan=np.tan(x_sin)
plt.scatter(x_sin, y_tan, color="b")
plt.title("Scatter_Plot")
plt.xlabel("X-Axies")
plt.ylabel("y-Axies")
plt.grid(True)
plt.show()
```

```python
y_tan=np.tan(x_sin)
plt.bar(x_sin, y_tan, color="g")
plt.title("Bar_plot")
plt.xlabel("X-Axies")
plt.ylabel("y-Axies")
plt.grid(True)
plt.show()
```

```python
y_tan=np.tan(x_sin)
plt.barh(x_sin, y_tan, color="b")
plt.title("Horizontal_Bar")
plt.xlabel("X-Axies")
plt.ylabel("y-Axies")
plt.grid(False)
plt.show()
```

```python
y_tan=np.tan(x_sin)
plt.scatter(x_sin, y_tan,marker="*",s=100, color="r")
plt.title("Scatter using Marker")
plt.xlabel("X-Axies")
plt.ylabel("y-Axies")
plt.grid(True)
plt.show()
```

```python
y_tan=np.tan(x_sin)
plt.plot(x_sin, y_tan, marker=".", color="b")
plt.title("Modern Graph")
plt.xlabel("X-Axies")
plt.ylabel("y-Axies")
plt.grid(True)
plt.show()
```

```python
y_tan=np.tan(x_sin)
plt.pie(y_tan)
plt.title("Pie_Chart")
plt.xlabel("****")
plt.ylabel("***")
plt.grid(True)
plt.show()
```
